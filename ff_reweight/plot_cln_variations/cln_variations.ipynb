```python
golden = (1 + 5 ** 0.5) / 2
figwidth = 6
fsize = 18
```

```python
def plot_cln_variations(filenames, labels, title):
    fig = plt.figure(figsize=(figwidth*golden,figwidth))
    ax = fig.gca()
    raw_data = []
    for f in filenames:
        raw_data.append(np.genfromtxt(f))
    for i,d in enumerate(raw_data):
        ax.plot(d[:,0], d[:,1], label = labels[i])
    # Shrink current axis by 20%
    box = ax.get_position()
    ax.set_position([box.x0, box.y0, box.width * 0.8, box.height])

    # Put a legend to the right of the current axis
    ax.legend(prop={'size':fsize}, loc='center left', bbox_to_anchor=(1, 0.5))
    ax.set_xlabel(r'$q^2$', fontsize=fsize)
    ax.set_ylabel(r'$d\Gamma/dq^2$ density', fontsize=fsize)
    ax.set_title(title, fontsize=fsize)
    ax.tick_params(length=8, width=1, labelsize=fsize)
    return fig
```

```python
_ = plot_cln_variations(['BDenu_CLN.csv', 'BDenu_CLN_drho_down.csv', 'BDenu_CLN_drho_up.csv'],
                        ['Default', r'$\rho^2$ down', r'$\rho^2$ up'],
                        r'$B\rightarrow{}De\nu$')
```
